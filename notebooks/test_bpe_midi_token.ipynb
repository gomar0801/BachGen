```python

from pathlib import Path
import json

# Our custom modules
from bachgen.display_and_play_partition import display_and_play
from bachgen.midi_bpe_token_style.music_xml_to_midi import music_xml_to_midi
from bachgen.midi_bpe_token_style.midi_to_music_xml import midi_to_music_xml
from bachgen.midi_bpe_token_style.midi_to_token import midi_to_token
from bachgen.midi_bpe_token_style.token_to_midi import token_to_midi
from bachgen.midi_bpe_token_style.display import display_midi_info, display_tokens

Path("temp").mkdir(exist_ok=True)
```

```python

print("\n" + "=" * 60)
print("PIPELINE 1: MusicXML → MIDI → MusicXML (No Tokenization)")
print("=" * 60)

# Step 1: Load and display original MusicXML
original_xml = "../musicxml_sample/full.musicxml"
print(f"\n📄 Step 1: Display original MusicXML")
display_and_play(original_xml, show_score=True, midi=False)

# Step 2: Convert to MIDI
print(f"\n🎵 Step 2: Convert MusicXML to MIDI")
midi_from_xml = music_xml_to_midi(original_xml, "temp/full_from_xml.mid")

# Step 3: Display MIDI info and visualization
print(f"\n🎹 Step 3: Analyze converted MIDI")
display_midi_info(midi_from_xml, show_notes=True)

# Step 4: Convert MIDI back to MusicXML
print(f"\n📄 Step 4: Convert MIDI back to MusicXML")
xml_from_midi = midi_to_music_xml(midi_from_xml, "temp/full_roundtrip.musicxml")

# Step 5: Display reconstructed MusicXML
print(f"\n📄 Step 5: Display reconstructed MusicXML")
display_and_play(str(xml_from_midi), show_score=True, midi=False)

print(f"\n✅ Pipeline 1 completed!")
print(f"Original:     {original_xml}")
print(f"MIDI:         {midi_from_xml}")
print(f"Reconstructed: {xml_from_midi}")
```

```python


print("\n" + "=" * 60)
print("PIPELINE 2: MusicXML → Tokens → MIDI → MusicXML (Full BPE)")
print("=" * 60)

# Step 1: Load original (same as before)
print(f"\n📄 Step 1: Original MusicXML")
print(f"File: {original_xml}")

# Step 2: Convert to MIDI (reuse from previous)
print(f"\n🎵 Step 2: Using MIDI from previous pipeline")
print(f"MIDI: {midi_from_xml}")

# Step 3: Convert MIDI to tokens
print(f"\n🔤 Step 3: Convert MIDI to tokens using pretrained REMI")
tokens, token_file = midi_to_token(midi_from_xml, "temp/minimal_tokens.json")

# Step 4: Display tokens
print(f"\n🔍 Step 4: Analyze tokens")
display_tokens(token_file, max_tokens=30)

# Step 5: Convert tokens back to MIDI
print(f"\n🎵 Step 5: Convert tokens back to MIDI")
midi_from_tokens = token_to_midi(token_file, "temp/minimal_from_tokens.mid")

# Step 6: Display reconstructed MIDI
print(f"\n🎹 Step 6: Analyze reconstructed MIDI")
display_midi_info(midi_from_tokens, show_notes=True)

# Step 7: Convert final MIDI to MusicXML
print(f"\n📄 Step 7: Convert final MIDI to MusicXML")
final_xml = midi_to_music_xml(midi_from_tokens, "temp/minimal_from_tokens.musicxml")

# Step 8: Display final result
print(f"\n📄 Step 8: Display final MusicXML")
display_and_play(str(final_xml), show_score=True, midi=False)

print(f"\n✅ Pipeline 2 completed!")
print(f"Original XML:   {original_xml}")
print(f"Intermediate MIDI: {midi_from_xml}")
print(f"Tokens:         {token_file}")
print(f"Reconstructed MIDI: {midi_from_tokens}")
print(f"Final XML:      {final_xml}")
```

```python
def compare_musicxml_files(original_path, reconstructed_path):
    """
    Compare two MusicXML files and show differences
    
    Args:
        original_path (str): Path to original MusicXML file
        reconstructed_path (str): Path to reconstructed MusicXML file
    """
    from music21 import converter
    
    print("🔍 MUSICXML COMPARISON")
    print("=" * 40)
    
    try:
        # Load both files
        print(f"Loading original: {original_path}")
        original = converter.parse(original_path)
        
        print(f"Loading reconstructed: {reconstructed_path}")
        reconstructed = converter.parse(reconstructed_path)
        
        # Basic comparison
        print(f"\n📊 BASIC COMPARISON:")
        print(f"  Duration:  {original.duration.quarterLength:.1f} → {reconstructed.duration.quarterLength:.1f} quarters")
        print(f"  Parts:     {len(original.parts)} → {len(reconstructed.parts)}")
        print(f"  Notes:     {len(original.flatten().notes)} → {len(reconstructed.flatten().notes)}")
        
        # Metadata comparison
        print(f"\n📝 METADATA COMPARISON:")
        orig_title = getattr(original.metadata, 'title', 'None') if hasattr(original, 'metadata') and original.metadata else 'None'
        recon_title = getattr(reconstructed.metadata, 'title', 'None') if hasattr(reconstructed, 'metadata') and reconstructed.metadata else 'None'
        
        orig_composer = getattr(original.metadata, 'composer', 'None') if hasattr(original, 'metadata') and original.metadata else 'None'
        recon_composer = getattr(reconstructed.metadata, 'composer', 'None') if hasattr(reconstructed, 'metadata') and reconstructed.metadata else 'None'
        
        print(f"  Title:     '{orig_title}' → '{recon_title}'")
        print(f"  Composer:  '{orig_composer}' → '{recon_composer}'")
        
        # Parts comparison
        print(f"\n🎼 PARTS COMPARISON:")
        for i in range(max(len(original.parts), len(reconstructed.parts))):
            if i < len(original.parts) and i < len(reconstructed.parts):
                orig_part = original.parts[i]
                recon_part = reconstructed.parts[i]
                
                orig_notes = len(orig_part.flatten().notes)
                recon_notes = len(recon_part.flatten().notes)
                
                print(f"  Part {i+1}:")
                print(f"    Notes:  {orig_notes} → {recon_notes}")
                print(f"    Name:   '{orig_part.partName}' → '{recon_part.partName}'")
            elif i < len(original.parts):
                print(f"  Part {i+1}: EXISTS → MISSING")
            else:
                print(f"  Part {i+1}: MISSING → EXISTS")
        
        # Musical elements
        print(f"\n🎵 MUSICAL ELEMENTS:")
        
        # Key signatures
        orig_keys = original.flatten().getElementsByClass('KeySignature')
        recon_keys = reconstructed.flatten().getElementsByClass('KeySignature')
        print(f"  Key signatures: {len(orig_keys)} → {len(recon_keys)}")
        if orig_keys:
            print(f"    Original: {[str(k) for k in orig_keys[:3]]}")
        if recon_keys:
            print(f"    Reconstructed: {[str(k) for k in recon_keys[:3]]}")
            
        # Time signatures
        orig_time = original.flatten().getElementsByClass('TimeSignature')
        recon_time = reconstructed.flatten().getElementsByClass('TimeSignature')
        print(f"  Time signatures: {len(orig_time)} → {len(recon_time)}")
        if orig_time:
            print(f"    Original: {[str(t) for t in orig_time[:3]]}")
        if recon_time:
            print(f"    Reconstructed: {[str(t) for t in recon_time[:3]]}")
        
        # Overall assessment
        note_diff = abs(len(original.flatten().notes) - len(reconstructed.flatten().notes))
        duration_diff = abs(original.duration.quarterLength - reconstructed.duration.quarterLength)
        
        print(f"\n✅ ASSESSMENT:")
        if note_diff <= 5 and duration_diff <= 1.0:
            print("  🟢 Excellent preservation - minimal differences")
        elif note_diff <= 20 and duration_diff <= 5.0:
            print("  🟡 Good preservation - some expected MIDI conversion loss")
        else:
            print("  🔴 Significant differences - check conversion process")
            
        print(f"  Note difference: {note_diff}")
        print(f"  Duration difference: {duration_diff:.2f} quarters")
        
    except Exception as e:
        print(f"❌ Error comparing files: {e}")

# Usage:
compare_musicxml_files("../musicxml_sample/full.musicxml", "temp/full_roundtrip.musicxml")
```
